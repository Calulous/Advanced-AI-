```python
# JAR SIZES MUST BE TAKEN FROM USER 
# SHOW THE PATH 
# even jar c shud be user input ... basically the user shud enter the size of 2 jugs and how much they want in the jug 

jugA_size = int(input("Enter size of jug A: "))
jugB_size = int(input("Enter size of jug B: "))

def water_jug_problem():
    jugA = 0  # current water in jug A
    jugB = 0  # current water in jug B
    
    print("Empty jugs right now:", "A:", jugA, "B:", jugB)
    
    # Step 1: Fill jug B
    jugB = jugB_size
    print("Fill jug B:", "A:", jugA, "B:", jugB)

    # Step 2: Pour B -> A # Step 3: transfering the 3L of water from jug 3 to jug 4 
                                         
    transfer = min(jugB, jugA_size - jugA) # checking the minimum value between jug3 & jug4 
                                           # check min(3,4-0) = 3 just an example
                                           # cannot just swap values as it does not take the sizes into account - allows for overflow
    jugA = jugA + transfer
    jugB = jugB - transfer
    print("Pour B -> A:", "A:", jugA, "B:", jugB)

    # Step 3: Fill jug B again
    jugB = jugB_size
    print("Fill jug B again:", "A:", jugA, "B:", jugB)

    # Step 4: Pour B -> A until A is full
    transfer = min(jugB, jugA_size - jugA)
    jugA = jugA + transfer
    jugB = jugB - transfer
    print("Pour again:", "A:", jugA, "B:", jugB)

    if jugA == 2 or jugB == 2:
        print("Finally! One jug has 2 liters.")
    else:
        print("This sequence did not reach 2 liters.")

water_jug_problem()

```
